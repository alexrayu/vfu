The general rule of thumb is that there are 4-5 article tags that will result in best SEO. Lower and higher amount will under-inform the search engine, or "scatter" the focus too much. Lets see how our sample dataset lives up to it.

```python
import pandas as pd
import seaborn as sns
from statsmodels.nonparametric.smoothers_lowess import lowess
import matplotlib.pyplot as plt

df = pd.read_csv('assets/article_tags_stat.csv')
sample = df.sample(10000)

filtered = lowess(sample.visits, sample.tags,
                  frac=.7, return_sorted=True)
plt.plot(filtered[:,0], filtered[:,1], 'tomato')

sns.regplot(sample.tags, sample.visits,
            lowess=True, scatter_kws={"s": 1})
```
